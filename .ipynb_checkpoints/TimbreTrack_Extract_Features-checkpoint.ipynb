```python
from comsar import TimbreTrack
from apollon.audio import AudioFile
import os
import numpy as np
import pandas as pd 
import glob
import os
from pathlib import Path
```

```python
#Instantiate Timbre Track
tt = TimbreTrack()
```

```python
#Extract Timbre features and write a .csv file for each sound file
files = glob.glob('FilePath/*.wav')
pitch = []
for file in files:
    a = tt.extract(file)
    p = Path(file)
    filecsv = Path(p.parent.as_posix() + '/' + p.stem + '.csv')
    a.features.to_csv(filecsv)
```
