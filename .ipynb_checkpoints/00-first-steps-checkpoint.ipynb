The command <it>str()</it> has converted the integer into a string. In the above example, the "999" is a dirty trick to ensure that all characters from the word are addressed, since the word "hallo" has clearly less than 999 characters. A better way to address all characters is the use of <font color="green">len</font>():

```python
word = "hallo"
word = word[0:1]+str(3)+word[2:len(word)]#len(word) is good practice
print(word)
```

### for, range, len

```python
x=4
for n in range(x):# this is a for loop. The indented code after the colon belongs to the loop.
    print(n)
#range means "take all integers < x "
```

```python
listing = [1,2,5,4,4] # this is a list, like a vector
for i in range(len(listing)):# here, len() is the length of the list
    print(listing[i])
```

```python
for i in range(-2,2):# we can also define any other integer to start with
    print(i)
```

```python
for i in range(-2,2,2):# and we can choose steps sizes other than 1
    print(i)
```

```python
for i in range(2,-3,-2):# we can also count backwards if we choose negative steps
    print(i)
```

### while, if, else, break

```python
i=2
while i < 6:# in contrast to for loops, while loope have no running variable
    print(i)
    i=i+1# therefore, the terminating condition must be met at any point through calculations inside the wile loop
```

```python
i=3
y = 7
while i < y:
    print(i)
    y = y-1 # here, the i stays unchanged, but the y changes, affecting the condition
```

```python
i = 3
while i < 6:
    print(i)
    if i <= 3:#if the current value of i is smaller than or equal to 3,
        i=i+1#it will be increased by 1 
    else:# otherwise:
        break# break terminates the loop immediately
```

Python knows the fundamentals of mathematics:

```python
x=2
y=3
print(x * y)#is a multiplication
print(x / y)#is a division
print(y + x)#is an addition
print(x - y)#is a subtraction
print(7.2 // 3)#is an integer division that does not output decimals (i.e., it gives the "floor" value of 7.2/3)
print(-11//2) # Here, the floor value of -5.5 is -6
print(7.2 % 3) # is the modulo function. It is the remainder of 7.2//3 (potentially including little errors)
print(divmod(11, 2)) # 
print(round(9.4))
print(2**3)# is 2 to the power of 3
print(pow(2, 3))# is also 2 to the power of 3
print(0**0) # is defined as 1
print(abs(c)) # gives the absolute, valid for integer, float, and complex
print(int(abs(c))) # converts a number into an integer (rounding)
print(float(3)) # converts integer into a float
print((2 + 3) * 5) # python knows the order of operators
c.conjugate() # is the conjugate of a complex number
```

# Your own function
Often, you define your own function that you want to use over and over. You need to write "def" and a function name with brackets (these may contain variables that the function expects), and a colon. Code that belongs to the function is indented. You may return a result.

```python
sr = 8000#sample rate
duration = 1#duration in seconds
def signal(freq, sr, dur):#here we define a new function with 3 inputs
    sign = (np.sin(2 * np.pi * np.arange(sr * duration) * freq / sr)).astype(np.float32)#!Note the indentation here!
    return sign#the function returns one value
# defining functions produces no output
```

```python
out = signal(1, sr, duration)#here we call the function and store the output in the variable "out"
print(out)
```

```python
def two(numerator,denominator):
    plussi = numerator + denominator
    minussi = numerator - denominator
    return plussi, minussi#To return two outputs, separate them by a comma
plu, minu = two(3,2)#and assign 2 variables to the function so that the first return is stored in plu and the secont, in minu
plu
```

<h1>Create a Plot</h1>

```python
import matplotlib.pyplot as plt#pyplot can be used to plot data

out2 = signal(-1, sr, duration)#here we call the function again and store the output in the variable "out2"
fig = plt.figure()#figure is our variable that will soon contain a figure. Right now you can think of it as an empty container.
plt.plot(out, color='blue', linestyle='dashed', label='sin')#plots the result
#besides 'dashed', you can use 'solid', 'dotted, 'dashdot'
plt.plot(out2, color='red', linewidth=7, label='-sin', marker="d")#this plots the result in the same plot
plt.grid(True)
plt.legend()#places a legend somewhere in the plot
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xticks(np.arange(0,sr*duration,1000))#range and interval of x-ticks
plt.yticks(np.arange(min(out),max(out)+.1,.2))#range and interval of y-ticks
plt.xlabel("Time [samples]", fontsize=16)#add a label for the x-axis
plt.ylabel("Elongation [mm]", fontsize=16)#add a label for the y-axis
plt.xlim(0, 6000)#crop the plot
plt.ylim(-1.01, 1.01);#the last line receives a semicolon to suppress the output
```
